```python
def timestamp_setup(df):
    df['timestamp'] = pd.to_datetime(df['match_date'] + ' ' + df['timestamp'])

    first_period_last_event = df.loc[df['period'] == 1, 'timestamp'].iloc[-1]

    df.loc[df['period'] == 2,'timestamp'] = df.loc[df['period'] == 2,'timestamp'] + datetime.timedelta(minutes=first_period_last_event.minute, seconds=first_period_last_event.second)

    df.set_index('timestamp', inplace=True)

    df.sort_index(inplace=True)

    return df
```

```python
def create_features(df, time_casa, time_fora):

    # Passes errados (casa x fora)
    df.loc[(df['pass_outcome_name'].isin(['Incomplete', 'Out', 'Unknown', 'Pass Offside'])) & (df['team'] == time_casa), 'pass_unsuccessful_casa'] = 1
    df.loc[(df['pass_outcome_name'].isin(['Incomplete', 'Out', 'Unknown', 'Pass Offside'])) & (df['team'] == time_fora), 'pass_unsuccessful_fora'] = 1   

    # Erros (casa x fora)
    df.loc[((df['type'].isin(['Dispossessed', 'Miscontrol', 'Error', 'Bad Behaviour', 'Offside', 'Dribbled Past'])) | (df['pass_unsuccessful_casa'] == 1)) 
           & (df['team'] == time_casa), 'errors_casa'] = 1
    
    df.loc[((df['type'].isin(['Dispossessed', 'Miscontrol', 'Error', 'Bad Behaviour', 'Offside', 'Dribbled Past'])) | (df['pass_unsuccessful_fora'] == 1)) 
           & (df['team'] == time_fora), 'errors_fora'] = 1

    # Passes na parte ofensiva do campo (casa x fora)
    df.loc[(df['type'] == 'Pass') & (df['team'] == time_casa) & (df['location'].apply(lambda v: isinstance(v, list) and v[0] >= 60)), 'pass_off_casa'] = 1
    df.loc[(df['type'] == 'Pass') & (df['team'] == time_fora) & (df['location'].apply(lambda v: isinstance(v, list) and v[0] >= 60)), 'pass_off_fora'] = 1

    # Passes na parte defensiva do campo (casa x fora)
    df.loc[(df['type'] == 'Pass') & (df['team'] == time_casa) & (df['location'].apply(lambda v: isinstance(v, list) and v[0] <= 60)), 'pass_def_casa'] = 1
    df.loc[(df['type'] == 'Pass') & (df['team'] == time_fora) & (df['location'].apply(lambda v: isinstance(v, list) and v[0] <= 60)), 'pass_def_fora'] = 1

    # Defensive Action na parte defensiva do campo (casa x fora)
    df.loc[(df['type'].isin(['Interception', 'Clearance', 'Foul Committed', 'Block', 'Duel', 'Pressure', 'Ball Recovery', 'Shield'])) & 
           (df['team'] == time_casa) & (df['location'].apply(lambda v: isinstance(v, list) and v[0] <= 60)), 'DA_casa'] = 1
    
    df.loc[(df['type'].isin(['Interception', 'Clearance', 'Foul Committed', 'Block', 'Duel', 'Pressure', 'Ball Recovery', 'Shield'])) & 
           (df['team'] == time_fora) & (df['location'].apply(lambda v: isinstance(v, list) and v[0] <= 60)), 'DA_fora'] = 1

    # Chutes feitos (casa x fora)
    df.loc[(df['type'] == 'Shot') & (df['team'] == time_casa), 'shot_casa'] = 1
    df.loc[(df['type'] == 'Shot') & (df['team'] == time_fora), 'shot_fora'] = 1

    # Gol feito (casa x fora)
    df.loc[(df['shot_outcome_name'] == 'Goal') & (df['team'] == time_casa), 'gol_casa'] = 1
    df.loc[(df['shot_outcome_name'] == 'Goal') & (df['team'] == time_fora), 'gol_fora'] = 1    

    return df
```

```python
match_ids = la_liga_2015_2016_matches['match_id']

df_matches = pd.DataFrame()

lag_size = 10

for match_id in match_ids:
    
    df = pd.DataFrame(json.load(open(f'../data/events/{match_id}.json' ,"rb")))

    normalized_dfs = [pd.json_normalize(df[col],sep='_').add_prefix(f'{col}_') for col in ['shot', 'duel', 'pass']]

    df = pd.concat([df.drop(columns=['shot', 'duel', 'pass'])] + normalized_dfs, axis=1)

    df['type'] = df['type'].apply(lambda k: k['name'])
    df['team'] = df['team'].apply(lambda k: k['name'])

    df['match_id'] = match_id
    df = df.merge(la_liga_2015_2016_matches, on='match_id', how='left')
    
    timestamp_setup(df)

    selected_columns = ['match_id', 'home_team', 'away_team', 'home_score', 'away_score', 'team', 'type', 'shot_outcome_name', 'duel_type_name', 'pass_outcome_name', 'location']

    df_new = df[selected_columns].copy()

    time_casa = df_new['home_team'].unique()[0]
    time_fora = df_new['away_team'].unique()[0]

    create_features(df_new, time_casa, time_fora)

    df_agg = temp_agg_df(df_new, 60)

    df_agg['lag_gol_fora'] = df_agg['gol_fora'].shift(-lag_size).fillna(0)
    df_agg['lag_gol_casa'] = df_agg['gol_casa'].shift(-lag_size).fillna(0)
    
    df_matches = pd.concat([df_matches, df_agg])

df_matches   
```
